```python
# import necessary libraries
from bs4 import BeautifulSoup
import requests
import re
import xlsxwriter
# function to extract html document from given url
def getHTMLdocument(url):
    # request for HTML document of given url
    response = requests.get(url)
    # response will be provided in JSON format
    return response.text

# assign URL
url_to_scrape = "https://www.onlinekhabar.com/"

# create document
html_document = getHTMLdocument(url_to_scrape)

# create soap object
links = BeautifulSoup(html_document, 'html.parser')

# create a workbook
workbook = xlsxwriter.Workbook('ok.xlsx')

# defining worksheet name
worksheet = workbook.add_worksheet("ok homepage url")
# defining excel header
worksheet.write(1,0,'Internal Links')
worksheet.write(1,1,'External Links')

# Start from the first cell. Rows and olumns are zero indexed.
row = 2
col = 0

# find all the anchor tags with "href" attribute starting with "https://"
InternalLinks = []
ExternalLinks = []
count = 0
for link in links.find_all('a', attrs={'href': re.compile("^https://")}):
    if link in links.find_all('a', attrs={'href': re.compile("^https://www.onlinekhabar.com")}):
        x = link.get('href')
        if x not in InternalLinks:
            InternalLinks.append(x)
            worksheet.write(row, col, x)
    else:
        y = link.get('href')
        if y not in ExternalLinks:
            ExternalLinks.append(y)
            worksheet.write(row, col + 1 , link.get('href'))
    count += 1
    row += 1
workbook.close()

for i in InternalLinks:    
    print('Internal Links :', i)
print('----------------------------------------------------------------')    
for j in ExternalLinks:
    print('External Links :', j)
print('----------------------------------------------------------------')  
print('Total Number of Internal Links:', len(InternalLinks))
print('Total number of External Links:', len(ExternalLinks))
print('Total Number of Links:', count)
# calculate repeated links
NumberofInternalLinks = (len(InternalLinks))
NumberofExternalLinks = (len(ExternalLinks))

repeatedLinks = count - ( NumberofInternalLinks + NumberofExternalLinks)
print("Number of Repeated Links:", repeatedLinks)

```
